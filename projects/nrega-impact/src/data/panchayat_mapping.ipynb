```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
project_dir = str(Path().resolve().parents[2])
nrega_dir = str(Path().resolve().parents[1])
nrega_dir
# raw_folder = os.path.join(
#     os.path.join(os.path.join(project_dir, "data"), "raw"), "receipt_budget"
# )
```

```python
lgd = pd.read_csv(os.path.join(nrega_dir, r"helper_files/ac_panchayat.csv"), encoding="iso-8859-1")
# renaming the concerned columns
# lgd = lgd.rename(columns={"stname": "state_name", "dtname": "district_name"})
# subsetting for necessary columns
lgd = lgd[["stname", "ST_CODE","dtname", "DT_CODE","gp_name","gp_code"]].drop_duplicates()
lgd.columns = ['state_name','state_code','district_name','district_code','gp_name','gp_code']
# uppercasing all names
# lgd["state_name_lower"] = lgd["state_name"].str.lower()
# lgd["district_name_lower"] = lgd["district_name"].str.lower()
# lgd["block_name"] = lgd["block_name"].str.upper()
lgd["gp_name_lower"] = lgd["gp_name"].str.lower()

# concatenating states, districts, blocks and panchayats

# lgd["state_dist_panch"] = lgd["state_name_lower"] + "_" + lgd["district_name_lower"] + "_" + lgd["gp_name_lower"]
up_lgd = lgd[lgd['state_name'] == 'UTTAR PRADESH'].copy()
up_lgd[['state_code','district_code']] = up_lgd[['state_code','district_code']].astype(int)
up_lgd
```

```python
der_lgd = pd.read_csv(os.path.join(nrega_dir, r"helper_files/St_Dt_Blk_Gp_LGD_Codes_20-5-21.csv")).dropna(subset=['gp_code'])
# # uppercasing all names
# der_lgd["state_name_lower"] = der_lgd["state_name"].str.lower()
# der_lgd["district_name_lower"] = der_lgd["district_name"].str.lower()
# der_lgd["block_name_lower"] = der_lgd["block_name"].str.lower()
der_lgd["gp_name_lower"] = der_lgd["gp_name"].str.lower().str.replace('rf ',';').str.replace('/',';')
der_lgd["gp_name_1"] = der_lgd["gp_name_lower"].str.lower().str.split(';').str[0].str.strip()
der_lgd["gp_name_2"] = der_lgd["gp_name_lower"].str.lower().str.split(';').str[1].str.strip()

# # concatenating states, districts, blocks and panchayats

# der_lgd["state_dist_panch"] = der_lgd["gp_name_lower"]
der_up_lgd = der_lgd[der_lgd['state_name'] == 'UTTAR PRADESH'].copy()
# up_lgd.drop(columns=['state_name','state_name_lower','district_name','district_name_lower','block_name','block_name_lower','gp_name','gp_name_lower'], inplace=True)
der_up_lgd.drop(columns=['state_name','district_name','block_name','gp_name'], inplace=True)
der_up_lgd[['state_code','district_code','block_code','gp_code']] = der_up_lgd[['state_code','district_code','block_code','gp_code']].astype(int)
der_up_lgd
```

```python
curr_state = "UTTAR PRADESH"
# print(der_up_lgd['gp_name_1'].nunique(), der_up_lgd['gp_name_2'].nunique())
```

```python
files = [file for file in  Path(nrega_dir).glob("./data/processed/block_lgd_mapped/*.csv") if file.stem == curr_state]
files
```

```python
df = pd.concat(pd.read_csv(f) for f in files).reset_index(drop=True)
df
```

```python
data_codes = df[['state','state_LGD_code','district','district_LGD_code','block_name','block_LGD_code','panchayat_name']].drop_duplicates()
# data_codes["state_name_lower"] = data_codes["state"].str.lower()
# data_codes["district_name_lower"] = data_codes["district"].str.lower()
# data_codes["block_name_lower"] = data_codes["block_name"].str.lower()
data_codes["gp_name_lower"] = data_codes["panchayat_name"].str.lower()
# concatenating states, districts, blocks and panchayats
# data_codes["state_dist_panch"] = data_codes["gp_name_lower"]
# data_codes.drop(columns=['state_name_lower','district_name_lower','gp_name_lower'], inplace=True)
data_codes
```

```python
merged_codes = data_codes.merge(der_up_lgd[['state_code','district_code','block_code','gp_code','gp_name_lower']], 
                                        left_on=['block_LGD_code','gp_name_lower'], right_on=['block_code','gp_name_lower'], how='left')

merged_codes = merged_codes.merge(der_up_lgd[['block_code','gp_code','gp_name_1']], 
                                    left_on=['block_LGD_code','gp_name_lower'], right_on=['block_code','gp_name_1'], how='left')
merged_codes['gp_code_x'] = merged_codes['gp_code_x'].fillna(merged_codes['gp_code_y'])
merged_codes = merged_codes.rename(columns={'gp_code_x':'gp_code'}).drop(['block_code_x','block_code_y','gp_code_y','gp_name_1'], axis=1)

merged_codes = merged_codes.merge(der_up_lgd[['block_code','gp_code','gp_name_2']], 
                                    left_on=['block_LGD_code','gp_name_lower'], right_on=['block_code','gp_name_2'], how='left')
merged_codes['gp_code_x'] = merged_codes['gp_code_x'].fillna(merged_codes['gp_code_y'])
merged_codes = merged_codes.rename(columns={'gp_code_x':'gp_code'}).drop(['block_code','gp_code_y','gp_name_2'], axis=1)

# merged_codes = merged_codes.merge(der_up_lgd, left_on=['block_LGD_code','gp_name_lower'], right_on=['block_code','gp_name_2'], how='left')
merged_codes
```
